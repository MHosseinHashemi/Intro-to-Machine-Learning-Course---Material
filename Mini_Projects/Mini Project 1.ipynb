## Strorytelling - Visualization

```python
import seaborn as sns
```

```python
sns.catplot(x="Type", kind="count", data=google_play)
```

```python
sns.displot(google_play, x="Rating", kind="kde" , hue ="Type" )
```

```python
sns.catplot(x="Rating", y="Category", kind="box", data=google_play , height =30 , aspect =0.6)
```

```python
sns.catplot(x="Price", y="Rating", kind="box", data=google_play , height =45 , aspect =2.5)
# the result is fine just open it in new tab and zoom in
```

```python
sns.relplot(x="Rating", y="Android Ver", data=google_play)
```

```python
sns.relplot(x="Rating", y="Reviews", data=google_play , height = 3.5 , aspect = 0.9 )
```

```python
sns.relplot(x="Rating", y="Installs", data=google_play)
```

```python
sns.relplot(x="Rating", y="Category", data=google_play)
```

```python
g = sns.FacetGrid(google_play, col="Category", height=5, aspect=.8)
g.map(sns.histplot, "Rating")
# its visible just open the image in new tab and see the result
```

```python
sns.catplot(x="Android Ver", y="Rating", kind="point", data=google_play , height = 5 , aspect = 8)
# its visible just open the image in new tab and see the result
```

```python
sns.catplot(x="Android Ver", y="Rating", kind="bar", data=google_play , hue = "Type" , height = 5 , aspect = 7)
# its visible just open the image in new tab and see the result
```
